```python
# Install pgmpy if not installed
# pip install pgmpy

from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.inference import VariableElimination
import pandas as pd


```

```python
# Step 1: Create dataset
data = pd.DataFrame({
    'Rain': ['No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No'],
    'TrafficJam': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No'],
    'ArriveLate': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No']
})

```

```python
# Step 2: Define Bayesian Network structure
model = DiscreteBayesianNetwork([
    ('Rain', 'TrafficJam'),
    ('TrafficJam', 'ArriveLate')
])

```

```python
# Step 3: Train the model
model.fit(data)

```

```python
# Step 4: Print CPDs
print("Conditional Probability Distributions:\n")
for cpd in model.get_cpds():
    print(cpd)
    print()

```

```python
# Step 5: Perform inference
inference = VariableElimination(model)

```

```python
# Step 6: Query
result = inference.query(
    variables=['ArriveLate'],
    evidence={'Rain': 'Yes'}
)

```

```python
# Step 7: Display result
print("Probability of ArriveLate given Rain = Yes:\n")
print(result)
```
